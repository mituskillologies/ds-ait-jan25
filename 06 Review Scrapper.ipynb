```python
import requests
from bs4 import BeautifulSoup

# Function to fetch reviews from a given URL
def fetch_reviews(url):
    # Send an HTTP GET request to the website
    response = requests.get(url)
    
    # If the request was successful (status code 200), parse the HTML content
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        text = soup.get_text(strip=False)
        return text
       
url = 'https://www.flipkart.com/portronics-sounddrum-1-10w-tws-portable-inbuilt-fm-type-c-charging-10-w-bluetooth-speaker/p/itm2e678c5363181'  # Replace with the actual URL of the product reviews page
text1 = fetch_reviews(url)
```
