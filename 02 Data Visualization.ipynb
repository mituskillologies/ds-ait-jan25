```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

##### Line Plot

```python
x = np.array([3,12,14,23,27,34,39,45,51,69])
y = np.array([10,23,16,34,45,59,67,71,83,95])
z = np.array([87,82,81,79,84,67,55,41,23,11])
p = np.array([45,33,56,67,87,65,38,64,58,65])
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, z)
```

```python
plt.plot(x, p)
```

```python
plt.title('My first plot')
plt.ylabel('Second Value')
plt.xlabel('First Value')
plt.grid()
plt.xlim(0,80)
plt.ylim(0,100)
plt.plot(x, y, color = 'r', marker = 'o')
for i in range(10):
    plt.annotate(xy=(x[i]-1, y[i]+2), text=i)
```

```python
plt.plot(x, y, label = '2022', color = 'r')
plt.plot(x, z, label = '2023', color = 'g')
plt.legend()
```

```python
plt.subplot(2,1,1)
plt.plot(x, y, label = '2022', color = 'r')
plt.subplot(2,1,2)
plt.plot(x, z, label = '2023', color = 'g')
```

```python
plt.figure(figsize=(16,9))
plt.plot(x, y, label = '2022', color = 'r')
plt.plot(x, z, label = '2023', color = 'g')
plt.legend()
```

##### Bubble Plot

```python
# Data
x = np.random.rand(50) * 100
y = np.random.rand(50) * 100
sizes = np.random.rand(50) * 1000  # Size of the bubbles

# Bubble Plot
plt.scatter(x, y, s=sizes, alpha=0.5, color='blue')
plt.title("Bubble Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

##### Box Plot

```python
# Data
data = [12, 15, 14, 10, 8, 9, 16, 17, 10, 13, 15, 14, 16, 18]

# Box Plot
sns.boxplot(data=data, color='skyblue')
plt.title("Box Plot Example")
plt.show()
```

##### Heat Map

```python
# Data: Create a random 10x10 matrix
data = np.random.rand(10, 10)

# Heat Map
sns.heatmap(data, annot=True, cmap="YlGnBu", cbar=True)
plt.title("Heat Map Example")
plt.show()
```

##### Violin Plot

```python
# Data: Using built-in seaborn dataset
tips = sns.load_dataset("tips")

# Violin Plot
sns.violinplot(x="day", y="total_bill", data=tips, palette="muted")
plt.title("Violin Plot Example")
plt.show()
```
